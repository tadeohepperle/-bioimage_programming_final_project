```python
# path = "data/sampleinput/FISH/12-09-12P0012201890-2.jpg"
pathes = [
    "data/sampleinput/FISH/12-09-12P0012201890-2.jpg",
#  "data/sampleinput/DISH1/125121A_S20200928_0014.jpg",
 ]

import numpy as np
import cv2
import matplotlib.pyplot as plt

for path in pathes:
    img = cv2.imread(path)
    img = cv2.blur(img, (5,5))
    # img[:,:,0] = cv2.equalizeHist(img[:,:,0])
    # img[:,:,1] = cv2.equalizeHist(img[:,:,1])
    # img[:,:,2] = cv2.equalizeHist(img[:,:,2])
    plt.imshow(img)
    plt.show()
    Z = img.reshape((-1,3))

    # plt.imshow(img)

    # # convert to np.float32
    Z = np.float32(Z)

    

    # # define criteria, number of clusters(K) and apply kmeans()
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    K = 4
    cen = np.array([[27.788949966430664, 25.383974075317383, 30.856565475463867], [30.886653900146484, 40.20023727416992, 46.6512451171875], [16.83529281616211, 19.812259674072266, 25.180030822753906], [44.52399444580078, 99.4610824584961, 56.579559326171875]])
    ret,label,center=cv2.kmeans(Z,K,cen,criteria,10,cv2.KMEANS_USE_INITIAL_LABELS, centers=cen)

    
    print(center.tolist())

    # # Now convert back into uint8, and make original image
    center = np.uint8(center)
    res = center[label.flatten()]
    res2 = res.reshape((img.shape))
    plt.imshow(res2)
    # cv2.imshow("res2", res2)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
```
